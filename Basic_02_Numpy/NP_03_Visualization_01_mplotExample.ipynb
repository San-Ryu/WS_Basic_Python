```python
## ★ Plot의 종류와 사용법

## ■ bar plot
## bar plot은 ax.bar와 같이 사용
## x : x축에 사용될 값
## height : 각 bar의 width
## bottom : y축의 base값
## align : 정렬 (center or edged)

## bar plot_일반
fig = plt.figure()
ax = fig.add_subplot()
ax.bar([0, 1, 2, 3, 4], [3, 2, 5, 4, 6])
plt.show()
```

```python
## bar plot_Width 조정
fig = plt.figure()
ax = fig.add_subplot()
ax.bar([0, 1, 2, 3, 4], [3, 2, 5, 4, 6], width = 0.5)
plt.show()
## 바의 너비가 감소한다. (Default = 0.8)
```

```python
## bar plot_Bottom 조정
fig = plt.figure()
ax = fig.add_subplot()
ax.bar([0, 1, 2, 3, 4], [3, 2, 5, 4, 6], bottom = 2)
plt.show()
## 최소값이 조정된다 (Default = 0)
```

```python
## bar plot_Align 조정
fig = plt.figure()
ax = fig.add_subplot()
ax.bar([0, 1, 2, 3, 4], [3, 2, 5, 4, 6], align = 'edge')
## ax.bar()
## 커서가 괄호 안에 위치했을 때, shift+tab키를 입력하면
## 관련 설명을 볼 수 있다. (Document 열람 가능)
plt.show()
## 바의 정렬을 설정한다
```

```python
## ■ histogram Plot
## histogram은 data의 분포를 나타낼 때 사용하며, 주로 다음과 같은 인자를 사용
## x : 입력 data
## bins : histogram을 나타낼 x축의 구간
##       (정수 입력시 구간의 갯수, list와 같은 sequence 입력시는 sequence 값이 구간으로 설정됨)
## range : x축 범위 (x.min, x.max)로 표현
## density : 확률 분포로 나타낼 것인지 여부

## histogram_일반
np.random.seed(1111)

fig = plt.figure()
ax = fig.add_subplot()
ax.hist(np.random.randn(500)) ## randn : 정규분포에서 값을 추출. (500개)
plt.show()
```

```python
## histogram_bin 입력
np.random.seed(1111)

fig = plt.figure()
ax = fig.add_subplot()
ax.hist(np.random.randn(500), bins = 50) 
## randn : 정규분포에서 값을 추출. (500개)
## bin : 몇개의 구간으로 표현할 것인가? -> 구간의 수를 늘려주면, 더욱 촘촘하게 나타낼 수 있다.
plt.show()
```

```python
## histogram_bin 입력
np.random.seed(1111)

fig = plt.figure()
ax = fig.add_subplot()

# ax.hist(np.random.randn(500), [-3, -2, -1, 0, 1, 2, 3]) 
## randn : 정규분포에서 값을 추출. (500개)
## bin : X좌표값을 대신 입력해줄 수 있다.
## 바로 위쪽의 bin 설정 그래프와 비슷하게 보인다!

ax.hist(np.random.randn(500), np.linspace(-2, 2, num = 40)) 
## 이렇게 표현해줄 수도 있다!
## np.linspace를 사용해서, 범위와 bin을 지정해줄 수 있다!


plt.show()
```

```python
## histogram_range 입력
np.random.seed(1111)

fig = plt.figure()
ax = fig.add_subplot()

ax.hist(np.random.randn(500), bins = 40, range = (-2, 2)) 

plt.show()
```

```python
## histogram_Density Option
np.random.seed(1111)

fig = plt.figure()
ax = fig.add_subplot()

ax.hist(np.random.randn(500), bins = 40, range = (-2, 2), density = True) 
## 해당 그래프가 확률 그래프가 되도록 Y축값을 바꿔준다.

plt.show()
```
